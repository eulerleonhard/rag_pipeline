```python
# 1. Imports and Spark Session Setup
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.types import StructType, StructField, DoubleType, StringType

import seaborn as sns
import matplotlib.pyplot as plt
from pyspark.sql.functions import col, isnan, when, count
```

```python
# Create a SparkSession. Ittells Spark how to access a cluster.
spark = SparkSession.builder \
    .appName("MLlib Iris") \
    .master("spark://spark-master:7077") \
    .config("spark.executor.memory", "512m") \
    .config("spark.executor.cores", "1") \
    .config("spark.ui.reverseProxy", "true") \
    .config("spark.ui.reverseProxyUrl", "http://localhost:4040") \
    .getOrCreate()

import os
print(os.environ.get("SPARK_MASTER"))
```

```python
print("Spark Master:", spark.sparkContext.master)
# print("Deploy Mode:", spark.sparkContext.deployMode)
print("Application ID:", spark.sparkContext.applicationId)
print("Spark version:", spark.version)
```

```python
# Load Iris CSV from HDFS
df = spark.read.csv("hdfs://hadoop-namenode:9000/data/Iris.csv", header=True, inferSchema=True)

# Preview
df.show(5)
```

```python
# Repartition to distribute across workers
df = df.repartition(6)

df.groupBy("Species").count().show()
```

```python
# Visualisation
pdf = df.toPandas()

sns.pairplot(pdf, hue="Species")
plt.show()
```

```python
# Check Nulls
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
# Index the label column ("Species")
indexer = StringIndexer(inputCol="Species", outputCol="label")
df_indexed = indexer.fit(df).transform(df)
df_indexed.tail(5)
```

```python
# Assemble features into vector
assembler = VectorAssembler(
    inputCols=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"],
    outputCol="features"
)
df_final = assembler.transform(df_indexed).select("features", "label")
df_final.count()
```

```python
# Split into training and test sets
train_data, test_data = df_final.randomSplit([0.8, 0.2], seed=123)

# Train Logistic Regression model
lr = LogisticRegression(maxIter=20, regParam=0.01)
model = lr.fit(train_data)

print("Coefficient matrix:\n", model.coefficientMatrix)
print("Intercept vector:\n", model.interceptVector)
```

```python
predictions = model.transform(test_data)
predictions.select("features", "label", "prediction", "probability").show(5, False)

from pyspark.ml.evaluation import MulticlassClassificationEvaluator

for metric in ["accuracy", "f1"]:
    evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName=metric)
    score = evaluator.evaluate(predictions)
    print(f"{metric.capitalize()}: {score:.4f}")
```
